```python
df.drop(['ID','LunchType','WklyStudyHours'],axis='columns',inplace=True)
```

```python
inputs = df.drop('WritingScore',axis='columns')
target = df.WritingScore
```

```python
from sklearn.preprocessing import LabelEncoder
le_Gender = LabelEncoder()
le_EthnicGroup = LabelEncoder()
le_ParentEduc = LabelEncoder()
le_TestPrep = LabelEncoder()
le_ParentMaritalStatus = LabelEncoder()
le_PracticeSport = LabelEncoder()
le_IsFirstChild = LabelEncoder()
le_TransportMeans = LabelEncoder()
```

```python
inputs['Gender_n'] = le_Gender.fit_transform(inputs['Gender'])
inputs['EthnicGroup_n'] = le_EthnicGroup.fit_transform(inputs['EthnicGroup'])
inputs['ParentEduc_n'] = le_ParentEduc.fit_transform(inputs['ParentEduc'])
inputs['TestPrep_n'] = le_TestPrep.fit_transform(inputs['TestPrep'])
inputs['ParentMaritalStatus_n'] = le_ParentMaritalStatus.fit_transform(inputs['ParentMaritalStatus'])
inputs['PracticeSport_n'] = le_PracticeSport.fit_transform(inputs['PracticeSport'])
inputs['IsFirstChild_n'] = le_IsFirstChild.fit_transform(inputs['IsFirstChild'])
inputs['TransportMeans_n'] = le_TransportMeans.fit_transform(inputs['TransportMeans'])
```

```python
inputs
```

```python
inputs.drop(['Gender', 'EthnicGroup', 'ParentEduc', 'TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'TransportMeans'], axis='columns', inplace=True)
```

```python
inputs
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
